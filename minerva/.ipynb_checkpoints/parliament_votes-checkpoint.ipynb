```python
from urllib.request import urlopen
import pandas as pd
from bs4 import BeautifulSoup
from dateutil.parser import parse

import matplotlib.pyplot as plt
from IPython.display import display 

plt.style.use('seaborn-paper')
%matplotlib inline
```

```python
link = 'https://www.sencanada.ca/en/senators-list/'
print('processing', link)
urlContent = urlopen(link).read().decode('utf-8')
```

```python


soup = BeautifulSoup(urlContent.encode('utf-8'), 'html.parser')
table = soup.find('table', id="senator-list-view-table")
# print(table)

# The first tr contains the field names.
headings = [th.get_text().strip(' \t\n\r') for th in table.find("tr").find_all("th")]

datasets = []
for row in table.find_all("tr")[1:]:
    dataset = [td.get_text().strip(' \t\n\r') for td in row.find_all("td")]
    datasets.append(dataset)

# print (datasets)
df = pd.DataFrame(datasets)
df.columns = headings
df.to_csv('data/senate_members.csv')
df.head()
```
